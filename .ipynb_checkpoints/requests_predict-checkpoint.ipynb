# Forecasting of Cosmetic Product Demand (Simulated Dataset In Tehran)
- This notebook applies a Prophet time-series model to forecast request volumes for the five highest-demand neighborhoods in Tehran.
# Steps
- Data preparation and time-series formatting
- Train Prophet model with linear trend + monthly seasonality
- 6 month forecasting
- evaluate accuracy with cross validation
- compare actual vs predicted trends in neighborhoods
# Outputs
- Forecast plots
- Cross validation metrics (RMSE, MAPE, coverage)
- Forecasting summary dataset

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from prophet import Prophet
from prophet.diagnostics import cross_validation
from prophet.diagnostics import performance_metrics
import logging
```

```python
df= pd.read_csv(r"./customers_sample.csv")
```

# prepare data for selected neighborhoods

```python
df["date"]=pd.to_datetime(df["date_"])
df["month"]=pd.to_datetime(df["date"]).dt.to_period('M').dt.to_timestamp()
```

```python
neighborhood_sums=df.groupby(['neighborhood','month']).agg(request_count=('address','count')).reset_index()
```

# Train model and predict

```python
all_forecasts = pd.DataFrame()
all_perf = pd.DataFrame()
top5_neighborhoods = neighborhood_sums.groupby('neighborhood')['request_count'].sum().sort_values(ascending=False).head(5).index

for neighborhood_name in top5_neighborhoods:
    df_sample = neighborhood_sums[neighborhood_sums['neighborhood'] == neighborhood_name][['month', 'request_count','neighborhood']].rename(columns={'month':'ds', 'request_count':'y','neighborhood':'neighborhood'}).reset_index()
    name_map = {
    'محله ونک': 'Vanak',
    'محله تهرانپارس غربی': 'West Tehranpars',
    'محله جنت آباد جنوبي': 'South Jannat Abad ',
    'محله سعادت آباد': 'Saadat Abad',
    'محله فردوس': 'Ferdows'}
    df_sample['neighborhood']=df_sample['neighborhood'].replace(name_map)
    new_neighborhood=df_sample['neighborhood'].iloc[0]
    df_sample['ds']=pd.to_datetime(df_sample['ds'])
    model = Prophet()
    model.add_seasonality(name='monthly', period=30.5, fourier_order=5)
    logging.getLogger("cmdstanpy").disabled = True
    model.fit(df_sample, algorithm='LBFGS')
    

    future = model.make_future_dataframe(periods=6, freq='MS')
    
    forecast = model.predict(future)
    df_cv = cross_validation(model, horizon='30 days', period='15 days', initial='120 days')
    perf = performance_metrics(df_cv)
    perf['neighborhood'] =new_neighborhood
    all_perf=pd.concat([all_perf,perf],ignore_index=True)
 
    forecast['neighborhood'] = new_neighborhood
   
    all_forecasts = pd.concat([all_forecasts, forecast], ignore_index=True)
    
    plot_df=forecast[['ds','yhat','neighborhood']]
    plot_df=plot_df.merge(df_sample[['ds','y']],on='ds',how='left')
    plt.figure(figsize=(12,6))
    sb.lineplot(data=plot_df, x="ds", y="y", marker="o", label=f"Actual_{new_neighborhood}_requests ")
    sb.lineplot(data=plot_df, x="ds", y="yhat", marker="o", label=f"Forecast_{new_neighborhood}_requests")
    plt.title("Trend of Requests - Actual vs Forecast")
    plt.xlabel("Month")
    plt.ylabel("Requests")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
all_perf
```

**Cross validation shows:**
- Ferdows and Saadat Abad: highly accurate (low MAPE) and reliable coverage
- West Tehranpars: moderate accuracy, reliable for longer horizons
- South Jannat Abad and Vanak: high errors and low coverage; forecasts are not reliable

```python
all_perf['horizon_days'] = all_perf['horizon'].astype(str).str.replace(' days','').astype(int)
heatmap_data=all_perf.pivot(index='neighborhood',columns='horizon_days',values='mape')
plt.figure(figsize=(10,6))
sb.heatmap(heatmap_data, annot=True, fmt=".2f", cmap='YlOrRd', cbar_kws={'label':'MAPE'})
plt.title("Heatmap of MAPE by Neighborhood and Horizon")
plt.ylabel("Neighborhood")
plt.xlabel("Horizon (days)")
plt.yticks(rotation=0)
plt.show()
```

# forcasting summary dataset for neighborhoods
- This dataframe includes predicted requests for the five neighborhoods along with key model performance metrics

```python
future_summary = all_forecasts[['ds', 'neighborhood', 'yhat', 'yhat_lower', 'yhat_upper']]
future_summary = future_summary[future_summary['ds'] > neighborhood_sums['month'].max()]

future_summary[['yhat', 'yhat_lower', 'yhat_upper']] = future_summary[['yhat', 'yhat_lower', 'yhat_upper']].round(0)
future_summary['MoM_%'] = future_summary.groupby('neighborhood')['yhat'].pct_change() * 100
future_summary['MoM_%'] = future_summary['MoM_%'].fillna(0).round(1)
future_summary['Unc_%'] = ( (future_summary['yhat_upper'] - future_summary['yhat']) / future_summary['yhat'] * 100 ).round(1)

final_summary=future_summary.rename(columns={
    'ds': 'Month',
    'yhat_lower': 'Plan_Low',
    'yhat': 'Plan_Base',
    'yhat_upper': 'Plan_High'
})
perf_summary = all_perf[['neighborhood','mae','mape','rmse']]
final_summary= final_summary.merge(perf_summary,on='neighborhood')
cols_to_round = ['Plan_Base','Plan_Low','Plan_High','MoM_%','Unc_%','mae','mape','rmse']
final_summary[cols_to_round] = final_summary[cols_to_round].round(1)
final_summary = final_summary.sort_values(['neighborhood','Month']).reset_index(drop=True)
```

```python
final_summary.to_csv(r"./forecasting_summary.csv")
```
